```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_circles 
from mpl_toolkits.mplot3d import Axes3D

from sklearn.model_selection import train_test_split
from sklearn import svm 
```

```python
#Make a large circle containing a smaller circle in 2d.
X, y = make_circles(n_samples = 600, noise = 0.03)

#show the data
plt.scatter(X[:, 0], X[:, 1], c = y, marker = '*') 
plt.show() 
```

## The above data is not linearly separable
## we need to convert to higher dimension

```python
#do transformation 2-D ->3D
X1 = X[:, 0].reshape((-1, 1)) 
X2 = X[:, 1].reshape((-1, 1)) 
X3 = (X1**2 + X2**2) 
X = np.hstack((X, X3)) 
```

```python
#show the data
fig = plt.figure(figsize=(10,10)) 
axes = fig.add_subplot(111, projection = '3d') 
axes.scatter(X1, X2, X1**2 + X2**2, c = y, depthshade = True) 
plt.show() 
```

## after converting 2-d to 3-d.
    - we use kernel=linear

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

svc = svm.SVC(kernel = 'linear') 
svc.fit(X_train, y_train) 

svc.score(X_test,y_test)
```
