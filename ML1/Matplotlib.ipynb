```python
# data visualization
# bar-plot, scatter-plots, histograms
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = x*2
y
```

```python
# Line Plot
```

```python
# plt.plot(x,y,color='g',linestyle=":",linewidth=2)
plt.plot(x,y)
plt.title('x and y')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
x1 = np.arange(1,11)
y1 = x1*2
y2 = x1*3
```

```python
plt.plot(x1,y1)
plt.plot(x1,y2)
plt.title('x and y')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1) # 1 row, 2 coloumn, 1= 1st one
plt.plot(x1,y1)
plt.subplot(1,2,2)
plt.plot(x1,y2)
plt.show()
```

```python
plt.subplot(2,1,1) # 1 row, 2 coloumn, 1= 1st one
plt.plot(x1,y1)
plt.subplot(2,1,2)
plt.plot(x1,y2)
plt.show()
```

```python
# Bar plot
```

```python
student = {'rbain':25,'raja':30,'susma':20}
names = list(student.keys())
values = list(student.values())
```

```python
# Vertical Bar
```

```python
plt.bar(names,values)
plt.title("bar plot")
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(False)
plt.show()
```

```python
# Horizontal Bar
```

```python
plt.barh(names,values)
plt.title("bar plot")
plt.xlabel('Names')
plt.ylabel('Marks')
plt.grid(False)
plt.show()
```

```python
## Scatter Plot
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [3,5,8,4,3,7,9,5,3]
b = [1,7,4,3,6,6,4,6,4]
c = [7,5,4,5,5,3,5,7,5]
```

```python
plt.scatter(x,a,marker='*',c='g',s=100)
plt.scatter(x,b)
plt.scatter(x,c)
plt.show()
```

```python
plt.subplot(1,3,1)
plt.scatter(x,a)
plt.subplot(1,3,2)
plt.scatter(x,b)
plt.subplot(1,3,3)
plt.plot(x,c)
```

```python
# Pie Chart
```

```python
fruits = ['apple','orange','Mango','guava']
quantity = [34,56,34,76]
```

```python
plt.pie(quantity,labels = fruits,autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(quantity,labels = fruits,autopct='%0.1f%%',radius = 2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
