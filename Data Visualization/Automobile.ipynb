```python
import pandas as pd
import seaborn as sns
```

```python
sns.set(color_codes=True)
```

```python
auto = pd.read_csv('Automobile_data.csv')
```

```python
auto.head()
```

```python
sns.displot(auto['stroke'])
```

```python
sns.displot(auto['engine-size'],kde=False,rug=True)
```

```python
sns.jointplot(data=auto, x='highway-mpg',y='city-mpg')
```

```python
# kind :{'scatter','reg','resid','kde','hex'}, optional
sns.jointplot(data=auto,x='highway-mpg',y='city-mpg',kind="reg")
```

```python
sns.pairplot(auto)
```

```python
sns.stripplot(data=auto, x="fuel-type", y="horsepower", jitter=True)
```

```python
sns.swarmplot(data=auto, x="fuel-type", y="horsepower")
```

```python
sns.boxplot( x="num-of-doors", y="highway-mpg",data=auto)
```

```python
sns.barplot(data=auto,x='body-style' ,y ='highway-mpg',hue='engine-location')
```

```python
sns.countplot(data=auto,x='body-style',hue='engine-location')
```

```python
sns.pointplot(data=auto,x='body-style',y='highway-mpg',hue='num-of-doors')
```

```python
# factorpot rename to catplot
sns.catplot(x='fuel-type',
              y='highway-mpg',
              hue='num-of-doors',
              col='engine-location',
              data=auto,
              kind='bar')

# kind='point','bar','count','box','violin','strip','swarm'
```

```python
# linear regression
sns.lmplot(x='highway-mpg',y='city-mpg',data=auto,hue='fuel-type')
```
