```python
import pandas as pd
```

```python
automobile = pd.read_csv('files/Automobile_data.csv')
```

```python
# check the head of hte DataFrame.
automobile.head()
```

```python
# How many rows and columns are there?
automobile.shape
```

```python
# What is the average Price of all cars in the dataset?

# automobile.apply("mean")
price = pd.to_numeric(automobile['price'],errors='coerce')
horsepower = pd.to_numeric(automobile['horsepower'],errors='coerce')
automobile['price'] = price
automobile['horsepower']=horsepower
price.mean()
```

```python
# what is the cheapest make and costliest make of car in the lot?
# automobile.columns
automobile.select_dtypes(include=['int64','float']).apply(['min','max'])['price']
```

```python
# How many cars have horsepower greater than 100?
# print(automobile.dtypes)
df1 = automobile.loc[automobile['horsepower']>100]
df1.shape
```

```python
# how many hatchback cars are in the dataset?
# print(automobile.dtypes)
df1 = automobile.loc[automobile['body-style']=='hatchback']
df1.shape
```

```python
# What are 3 moost commonly found cars in the dataset?
# automobile.columns
automobile["body-style"].value_counts().head(n=3)

```

```python
# Someone purchased a car for 7099, what is the make of the car?
automobile.loc[automobile['price']==7099]['make']
```

```python
# which cars are priced greater than 40000?
automobile.loc[automobile['price']>40000]
```

```python
# Which are the cars that are both a sedan and priced less that 7000?
automobile.loc[(automobile['price']<7000) & automobile['body-style'].isin(['sedan'])]
```
